# Plot Parameter Distributions (xHI and fX)

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the saved dataset
data_path = "/user1/supriyo/ml_project/SBI_step_by_step/UGMRT_500h/Data/dataset.pkl"
with open(data_path, "rb") as f:
    data = pickle.load(f)

theta = data["theta"]
x_1d = data["x_1d"]
x_2d = data["x_2d"]

print(f"✅ Data loaded successfully!")
print(f"theta shape: {theta.shape}")
print(f"x_1d shape: {x_1d.shape}")
print(f"x_2d shape: {x_2d.shape}")
print(f"\nParameters (xHI, fX):")
print(f"xHI range: [{theta[:, 0].min():.4f}, {theta[:, 0].max():.4f}]")
print(f"fX range:  [{theta[:, 1].min():.4f}, {theta[:, 1].max():.4f}]")
```

## Plot 1: Histogram of xHI (Neutral Fraction)

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(theta[:, 0], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
plt.xlabel('xHI (Neutral Fraction)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of xHI', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(theta[:, 1], bins=30, color='coral', edgecolor='black', alpha=0.7)
plt.xlabel('fX (log10 X-ray heating)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of fX', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)

plt.tight_layout()
#plt.savefig('/user1/supriyo/ml_project/SBI_step_by_step/UGMRT_500h/Outputs/param_histograms.png', dpi=150, bbox_inches='tight')
plt.show()

print("✅ Saved: param_histograms.png")
```

## Plot 2: 2D Scatter Plot (xHI vs fX)

```python
plt.figure(figsize=(8, 6))
plt.scatter(theta[:, 0], theta[:, 1], alpha=0.6, s=50, color='purple', edgecolors='black', linewidth=0.5)
plt.xlabel('xHI (Neutral Fraction)', fontsize=12)
plt.ylabel('fX (log10 X-ray heating)', fontsize=12)
plt.title('Parameter Space: xHI vs fX', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
#plt.savefig('/user1/supriyo/ml_project/SBI_step_by_step/UGMRT_500h/Outputs/param_scatter.png', dpi=150, bbox_inches='tight')
plt.show()

print("✅ Saved: param_scatter.png")
```

## Plot 3: 2D Hexbin Plot (Density visualization)

```python
plt.figure(figsize=(8, 6))
hexbin = plt.hexbin(theta[:, 0], theta[:, 1], gridsize=20, cmap='YlOrRd', mincnt=1)
plt.xlabel('xHI (Neutral Fraction)', fontsize=12)
plt.ylabel('fX (log10 X-ray heating)', fontsize=12)
plt.title('Parameter Space Density: xHI vs fX', fontsize=14, fontweight='bold')
cbar = plt.colorbar(hexbin)
cbar.set_label('Count', fontsize=11)
plt.tight_layout()
# plt.savefig('/user1/supriyo/ml_project/SBI_step_by_step/UGMRT_500h/Outputs/param_hexbin.png', dpi=150, bbox_inches='tight')
plt.show()

print("✅ Saved: param_hexbin.png")
```
